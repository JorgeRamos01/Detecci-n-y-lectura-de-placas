```python
import numpy as np

def levenshtein(seq1, seq2):  
    size_x = len(seq1) + 1
    size_y = len(seq2) + 1
    matrix = np.zeros ((size_x, size_y))
    for x in range(size_x):
        matrix [x, 0] = x
    for y in range(size_y):
        matrix [0, y] = y

    for x in range(1, size_x):
        for y in range(1, size_y):
            if seq1[x-1] == seq2[y-1]:
                matrix [x,y] = min(
                    matrix[x-1, y] + 1,
                    matrix[x-1, y-1],
                    matrix[x, y-1] + 1
                )
            else:
                matrix [x,y] = min(
                    matrix[x-1,y] + 1,
                    matrix[x-1,y-1] + 1,
                    matrix[x,y-1] + 1
                )
    #print (matrix)
    return (matrix[size_x - 1, size_y - 1])
```

```python
lista_tc=[]
for i in range(len(deteccion_placas_etiquetado.placa.values)):
    verdadera = deteccion_placas_etiquetado.loc[i,"texto"]
    dist_minima=100
    token_min=""
    #print(verdadera)
    for l in range(num_logs):
        toks =lista_logs[l][i].split(":")[1].strip().split(",") 
        #print(toks)
        for palabra in toks:
            dist_lev = levenshtein(verdadera,palabra)
            #print(dist_lev)
            if(dist_lev<dist_minima):
                dist_minima = dist_lev
                token_min=palabra
    if(token_min is None):
        lista_tc.append((verdadera,"",dist_minima))
    else:
        lista_tc.append((verdadera,token_min,dist_minima))
```
