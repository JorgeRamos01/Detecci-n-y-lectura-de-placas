# Usando Tesseract para detectar caracteres en las placas

### Importamos las librerias necesarias

```python
import sys
import numpy as np
import cv2
import os
import re
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt
```

### Cargamos la imagen de la placa y la mostramos

```python
img = cv2.imread("messigray.png")            # read in training numbers image
plt.imshow(img, interpolation="bicubic")
```

### Generamos un borde blanco alrededor de la imagen

```python
row, col= img.shape[:2]
bottom= img[row-2:row, 0:col]
mean= cv2.mean(bottom)[0]

bordersize=10
border=cv2.copyMakeBorder(img, top=bordersize, bottom=bordersize, left=bordersize, right=bordersize, borderType= cv2.BORDER_CONSTANT, value=[255,255,255] )
plt.imshow(border, interpolation="bicubic")
```

### Aplicamos Tesseract a la imagen de la placa y mostramos el texto extraido

```python
text=pytesseract.image_to_string(border)
text=re.sub(r'[^\w]', ' ', text)  #Removemos caracteres especiales
print(text)
```
